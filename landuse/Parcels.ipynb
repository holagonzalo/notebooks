SANGIS publishes a geographic file that links basic tax assessment data to parcels in San Diego. The file appears to be comprehensive, with about 1M parcel records. The data in the file includes fields for things like:
* identification numbers
* Owner names and addresses
* Legal name, for businesses
* Several different types of land use and zoning codes
* Tax assessment value for land and impovements

While the file is regularly updated -- it appears to be updated approximately monthly -- the documentation notes:

> Parcel Zone and Land Use codes are used by County appraisers. They are 
> not well maintained and are not considered reliable for other purposes

There are other files that are more reliable for this information, but this file should be sufficient for "order of magnitude" querys about how value and land use are connected. 

This file uses a lot of codes, which are defined in the [documentation on the SANGIS website](http://www.sangis.org/download/index.html). 


```python
%matplotlib inline
import ambry
import pandas as pd
l = ambry.ilibrary()
p = l.get("sangis.org-parcels-orig-parcels-geo").partition
```

```python
df = p.select("SELECT own_name1, legldesc, asr_zone, asr_landus, asr_total, nucleus_zo, nucleus_us  FROM parcels").pandas
```

```python
df.count()
```

## Land Use Codes and Assessed Value

```python
x = df[['nucleus_us','asr_total','asr_zone', 'asr_landus' ]].groupby(df.nucleus_us)
sums = x.sum().sort('asr_total', ascending=False)
sums
counts = x.count()['asr_total']
df2 = pd.DataFrame(index=sums.index)
df2['sums'] = sums
df2['counts'] = counts
df2['avg'] = df2['sums'] / df2['counts']
df2.sort('counts', ascending=False)
```
